```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, cophenet, dendrogram
from scipy.spatial.distance import pdist
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.cluster import AgglomerativeClustering, KMeans
import os

# Supongamos que 'data' contiene tus datos
data = pd.read_csv('Mall_Customers.csv')

# Seleccionar las características de interés
X = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].values

# Ajuste del rango de clusters para KMeans
Nc_kmeans = range(1, len(X) + 1)
kmeans = [KMeans(n_clusters=i, n_init=10) for i in Nc_kmeans]
score_kmeans = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]

# Graficar la puntuación en función del número de clusters para KMeans
plt.plot(Nc_kmeans, score_kmeans, marker='o')
plt.xlabel('Número de clusters')
plt.ylabel('Puntuación')
plt.title('Método del codo para determinar el número óptimo de clusters (KMeans)')
plt.show()

# Calcular el coeficiente de correlación cofenética
X_dist = pdist(X)
Z = linkage(X, 'ward')
coef_cofenetico, _ = cophenet(Z, X_dist)
print("Coeficiente de correlación cofenética:", coef_cofenetico)

# Calcular medidas de calidad para Clustering Jerárquico
hier_clustering = linkage(X, method="ward", metric="euclidean")
labels_hier = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward').fit_predict(X)
print("\nMedidas de calidad para Clustering Jerárquico:")
print("Puntaje de silueta:", silhouette_score(X, labels_hier))
print("Índice de Calinski-Harabasz:", calinski_harabasz_score(X, labels_hier))
print("Davies-Bouldin Index:", davies_bouldin_score(X, labels_hier))

# Evitar la advertencia sobre fuga de memoria en Windows con MKL
os.environ['OMP_NUM_THREADS'] = '1'
```

```python
# Asumiendo que 'sepal_length' y 'sepal_width' corresponden a 'Age' y 'Annual Income (k$)'
def plot_dendrogram(model, **kwargs):
    '''
    This function extracts the information from an AgglomerativeClustering model
    and plots its dendrogram with the dendrogram function from scipy.cluster.hierarchy
    '''
    from scipy.cluster.hierarchy import dendrogram
    import numpy as np    
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count
    distances = np.arange(model.children_.shape[0])
    linkage_matrix = np.column_stack([model.children_, distances, counts]).astype(float)
    # Plot
    plt.figure(figsize=(20, 7))
    dendrogram(linkage_matrix, **kwargs)
    plt.title('Dendrograma')
    plt.xlabel('Sample index')
    plt.ylabel('Euclidean distance')
    plt.show()

plot_dendrogram(hier_clustering)
```

```python
# Deshabilitar las advertencias de FutureWarning
warnings.filterwarnings("ignore", category=FutureWarning)
# Lee los datos desde un archivo CSV
data = pd.read_csv('Mall_Customers.csv')

# Asumiendo que 'data' es un DataFrame de pandas
X = data[['CustomerID', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']].values

# Ajuste del rango de clusters
range_n_clusters = range(2, min(15, len(X)))

valores_medios_silhouette = []
for n_clusters in range_n_clusters:
    # Definir y ajustar el modelo de clustering
    modelo = AgglomerativeClustering(
        affinity='euclidean',
        linkage='ward',
        n_clusters=n_clusters
    )
    cluster_labels = modelo.fit_predict(X)
    
    if len(np.unique(cluster_labels)) > 1:
        # Calcular el índice silhouette para cada cluster
        silhouette_avg = silhouette_score(X, cluster_labels)
        valores_medios_silhouette.append(silhouette_avg)

# Graficar la evolución de la media de los índices silhouette
plt.plot(range_n_clusters, valores_medios_silhouette, marker='o')
plt.title("Evolución de media de los índices silhouette")
plt.xlabel('Número clusters')
plt.ylabel('Media índices silhouette')
plt.show()
```
